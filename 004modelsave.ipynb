```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import RMSprop
```

```python
times = [2,4,6,8]
lession= [0,4,2,3]
sungjuk = [81,93,91,97]
```

```python
data = [[2,0,81],[4,4,93],[6,2,91],[8,3,97]]
xdata1 = [x[0] for x in data]
xdata2 = [x[1] for x in data]
ydata = [x[2] for x in data]
```

```python
a = tf.Variable(tf.random_uniform([1], 0, 10,
                dtype=tf.float64, seed=1))
b = tf.Variable(tf.random_uniform([1], 0, 10,
                dtype=tf.float64, seed=1))
c = tf.Variable(tf.random_uniform([1], 0, 100,
                dtype=tf.float64, seed=1))
```

```python
y = a * xdata1 + b * xdata2 + c
```

```python
rmse = tf.sqrt(tf.reduce_mean(tf.square(y - ydata)))
```

```python
learn_rate = 0.1
```

```python
gradient_decent = tf.train.GradientDescentOptimizer\
                (learn_rate).minimize(rmse)
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for step in range(2001):
        sess.run(gradient_decent)
        if step % 10 == 0:
            print('훈련횟수 epoch %.f, RMSE %.3f,'\
            '기울기1 %.3f, 기울기2 %.3f, 절편 %.3f' %
                  ( step, sess.run(rmse),
                    sess.run(a), sess.run(b),
                    sess.run(c) ))
```

```python
from tensorflow.keras import utils
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import load_model
import pandas as pd
```

```python
iris = pd.read_csv('data/iris.csv')
```

```python
iris = pd.read_csv('data/iris.csv', header=None)

x_data = iris.iloc[:, 0:4]
y_data = iris.iloc[:, 4]


le = LabelEncoder()
le.fit(y_data)
y_data = le.transform(y_data)
y_data = utils.to_categorical(y_data)
```

#### 시드 설정

```python
seed = 211202
np.random.seed(seed)
tf.set_random_seed(seed)
```

#### 신경망, 손실함수, 옵티마이저 설정

```python
model = Sequential()
model.add(Dense(10, input_dim=4, activation='relu'))
model.add(Dense(3, activation='softmax'))

model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy', 'mse'])
```

```python
model.fit(x_data, y_data, epochs=1000, verbose=0)
```

```python
print('%.4f' % model.evaluate(x_data, y_data)[1])
```

#### 모델 저장 : 객체.save(경로)

```python
model.save('models/iris.h5') # 모델 확장자 h5
```

### 과적합 문제 해결
+ 1998년 존스홉킨스 대학 세즈노프스키 교수가 힌튼 교수가 발표한 역전파 알고리즘에 과심을 가짐

+ 즉, 은닉층과 역전파의 효과가 얼마나 큰지 궁금

+ 따라서, 광석과 일반돌에 음파를 쏘아 측정치 조사

+ 과연 이 돌을 잘 분류할 수 있을지 실험(sonar.csv)

#### 논문결과 : 은닉층의 수가 올라갈수록 데이터 예측 정확도도 같이 상승
+  학습/검증 데이터로 분리해서 진행하면 증가속도가 다소 지연됨

#### 과적합 방지 방법 - 데이터를 train/test로 분리해서 훈련 실시 (+ 교차검증)

```python
sonar = pd.read_csv('data/sonar.csv', header=None)
```

```python
import numpy as np
```

```python
x_data = sonar.iloc[:, 0:60]
y_data = sonar.iloc[:, 60]
print(y_data[:5])
```

```python
# 전처리 수행 - 범주형 데이터 변환

le = LabelEncoder()
y_data = le.fit_transform(y_data)
print(y_data[:5])
```

```python
# 난수 지정
seed = 190121
np.random.seed(seed)
tf.set_random_seed(seed)
```

```python
# 신경망 구성 : 
# 특성수 : 60
# 입력노드 : ???, 활성화 함수 : relu
# 최종 출력노드 : 1, 활성화 함수 : sigmoid
model = Sequential()
```

#### 은닉층 설정 예시

```python
model.add(Dense(40, input_dim=60, activation='relu')) # 입력노드 60개를 40으로 줄임
```

```python
model.add(Dense(20, input_dim=40, activation='relu')) # 은닉층 : 40개를 20으로 줄임
```

```python
model.add(Dense(1, activation='sigmoid')) # 최종 노드 1개
```

```python
model.add(Dense(25, input_dim=60, activation='relu'))
```

```python
model.add(Dense(1, activation='sigmoid'))
```

```python
# 손실함수, 최적화함수 정의
model.compile(loss='binary_crossentropy',  # 이항분류 - 둘 중 하나 분류
              optimizer='adam', # RMSPROP + MOMENTUM
              metrics=['accuracy','mse']) # 가중치 설정 확인용 정확도와 오차
```

```python
model.fit(x_data, y_data, epochs=1000, verbose=0)
```

```python
print(model.evaluate(x_data, y_data)[1]) # 결과 1.0 과적합 발생 가능성 높음
```

```python
from sklearn.model_selection import train_test_split
```

```python
# 과적합을 방지하기 위해
# 데이터셋에서 train/test 데이터로 분리
xtrain, xtest, ytrain, ytest = \
     train_test_split(x_data, y_data,
                      test_size=0.3,
                      random_state=seed)

model.fit(xtrain, ytrain, epochs=1000, verbose=0)

print(model.evaluate(xtest, ytest)[1])
```
